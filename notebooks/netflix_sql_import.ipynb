```python
import pandas as pd
from sqlalchemy import create_engine

# Load CSV
df = pd.read_csv(r"E:\data Analystic\Netflix\data\raw\netflix_titles.csv")

# Rename cast column
df.rename(columns={'cast': 'cast_members'}, inplace=True)

# SQL Server connection
server = 'RADHE\\SQLEXPRESS'
database = 'NetflixPortfolioDB'

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

engine = create_engine(connection_string)

# Upload dataframe to SQL Server
df.to_sql(
    name='netflix_titles',
    con=engine,
    if_exists='replace',
    index=False
)

print("Data imported successfully!")
```
